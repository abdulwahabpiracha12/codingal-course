```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset('iris')
print(iris.head())
```

```python
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal length vs Sepal width")
plt.show()
```

```python
sns.lineplot(data=iris, x="sepal_length", y="petal_width", hue="species")
plt.title("Sepal length vs Petal width")
plt.show()
```

```python
sns.histplot(data=iris, x="petal_length", hue="species", kde="true")
plt.title("Distribution of Petal length")
plt.show()
```

```python
sns.kdeplot(data=iris, x="sepal_width", hue="species", fill="true")
plt.title("KDE Plot of Sepal Width")
plt.show()
```

```python
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Box plot of petal Length by Species")
plt.show()
```

```python
sns.voilinplot(data=iris, x="species", y="petal_width")
plt.title("Voilin Plot of Petal Width")
plt.show()
```

```python
sns.pairplot(iris, hue = 'species')
plt.show()
```
